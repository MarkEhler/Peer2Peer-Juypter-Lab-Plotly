### Correlation
<ul>
    
<li>Find the mean of all the x-values and y-values.</li>
<li>Find the standard deviation of all the x-values (call it sx) and the standard deviation of all the y-values (call it sy). </li>
<li>For each of the n pairs (x, y) in the data set, subtract their respective means.</li>
<li>Add up the n results from Step 3.</li>
<li>Divide the sum by sx ∗ sy.</li>

$$Correlation = \sum\frac{\text{(x - mean_x) * (y - mean_y)}}{\text{sx * sy}}$$

### Heat Map

```python
import plotly.figure_factory as ff

corrs = heatmap.iloc[:,:].corr()

figure = ff.create_annotated_heatmap(
    z=corrs.values,
    x=list(corrs.columns),
    y=list(corrs.index),
    colorscale='Portland',
    annotation_text=corrs.round(2).values,
    showscale=True, reversescale=True)

figure.layout.margin = dict(l=200, t=200)
figure.layout.height = 800
figure.layout.width = 1000

iplot(figure)
```
